```python
plt.style.use('default')
plt.rc('text',usetex=True)
plt.rc('lines', linewidth=1)
```

```python
xb = 9./13
xc = 6./13
```

```python
epsMin = 1e-10
epsMax = .2
epsVal = np.linspace(epsMin, epsMax)
entVal1 = []
entVal2 = []
```

```python
for ep in epsVal:
    entVal1 += [entropy(probVec1(ep, xb))]
    entVal2 += [entropy(probVec2(ep, xc))]
```

```python
entVal1 = np.array(entVal1)
entVal2 = np.array(entVal2)
```

```python
fntSz = 20
fntSz1 = 14
fig, ax1 = plt.subplots()

ax1.set_xlim(epsMin,epsMax)

ax1.plot(epsVal, entVal1-entVal2,c='k',label=r'$x_b = 3/4$')
ax1.plot(epsVal, 0*epsVal,c='k',linestyle='--')

ax1.set_ylabel(r'$\Delta(\epsilon)$',labelpad=10,size=fntSz,rotation='horizontal')
ax1.set_xlabel(r'$\epsilon$',labelpad=10,size=fntSz)

ax1.tick_params(direction='out',top=0,right=0,labelsize=fntSz1)
ax1.tick_params(axis='y',labelright=0,labelleft=1)

savefig('Fig3LSDiff.pdf',dpi=400,bbox_inches='tight',transparent=True)
```
